```python
#conjunto de chave e valor
eu = {"nome": "Erick", "idade": 26, "cargo":"Analista Desenvolvedor"}
print(eu)
```

```python
#set é um conjunto não ordenado e não repetido de elementos
animais  =  {"gato","cachorro","papagaio"}
animais.add("cachorro")
print(animais)
```

```python
#tupla é um conjunto imutavel de elementos
tp = (1,2,6,5)
print(tp) 
```

```python
#teste lógico
print("saldo" in eu)
print("nome" in eu)
```
