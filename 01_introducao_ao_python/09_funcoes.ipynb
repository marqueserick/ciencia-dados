```python
#funcao com retorno
def potencia(n,m):
  return n**m
```

```python
print(potencia(2,3))
```

```python
#funcao com valor default
def potencia_default(n,m=2):
  return n**m
```

```python
print(potencia_default(5))
```

```python
'''
funcoes do pyhton:
abs() - retorna valor absoluto
max() - retorna o maior valor de uma lista
min() - retorna menor valor de uma lista
upper() - retorna texto com todas letras maiusculas
entre outras
'''
print(abs(-200))
lst = [45,1,23,46,23]
print(lst)
print(f"Maior valor da lista: {max(lst)}")
print(f"Menor valor da lista: {min(lst)}")
print(f"Soma dos valores: {sum(lst)}")
print(f"Arredondamento {round(2.345678,2)}")
```

```python
from statistics import *
print(f"Média dos valores: {mean(lst)}")
print(f"Mediana dos valores: {median(lst)}")
print(f"Moda: {mode(lst)}")
print(f"Desvio padrão dos valores: {stdev(lst)}")
print(f"Variância: {variance(lst)}")
```
