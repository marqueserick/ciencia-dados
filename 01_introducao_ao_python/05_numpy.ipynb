```python
#facilita operaçoes matematicas
#objeto de matriz multidimensional
#possui tamanho fixo e todos elementos devem ter o mesmo tipo
import numpy as np
a = np.array([12,36,5,7,5])
print(a)
print(a.dtype)
```

```python
#cria uma matriz unidimensional
mt = np.array([12,13,134,15])
print(mt)
print(type(mt))
```

```python
#criar um array com tipo especificico
mt_float = np.array([1,2,3], dtype = np.float64)
print(mt_float)
print(type(mt_float))
mt_int = np.array([1,2,3,4], dtype = np.int32)
print(mt_int)
print(type(mt_int))
```

```python
#converter dados do array
mt = np.array([2.33, 1.9, 9.19999])
print(mt)
mt_int = mt.astype(np.int32)
#ao fazer a conversão, os valores são truncados
print(mt_int)

#podemos fazer o inverso
mt_int = np.array([1,2,3])
print(mt_int)
mt_float = mt_int.astype(np.float64)
print(mt_float)
```

```python
#array bidimensional
mt_bidimensional = np.array([[1,2,3],[2,3,4],[3,4,5]])
print(mt_bidimensional)
```

```python
#criar arrays vazias tipificadas
#empty significa que não são inicializadas, não que são vazias
vazio = np.empty([3,2], dtype = np.int32)
print(vazio, end="\n\n")
zeros = np.zeros([4,3])
print(zeros, end="\n\n")
um = np.ones([5,3])
print(um, end="\n\n")
diagonal = np.eye(5)
print(diagonal, end="\n\n")
```

```python
#gerar matriz com valores aleatorios entre 0 e 1
aleatorios = np.random.random((5))
print(aleatorios, end="\n\n")
#valores aleatórios distribuição normal com numéros negativos
aleatorios2 = np.random.randn((5))
print(aleatorios2, end="\n\n")
#valores aleatorios 3x4
aleatorios3 = np.random.random((3,4))
print(aleatorios3, end="\n\n")
#multiplicando valores aleatorios
aleatorio4 = (10*np.random.random((3,4)))
print(aleatorio4)
```

```python
#outra forma de gerar aleatorio
#uso de semente
gnr = np.random.default_rng(1)
print(gnr, end="\n\n")
aleatorio5 = gnr.random(5)
print(aleatorio5)
#gerar inteiros
aleatorio6 = gnr.integers(20, size=(4,3))
print(aleatorio6)
```

```python
#unique remove repetições e ordena itens
j = np.array([11,12,13,14,1,12,13,5,6,8,9,11,12,11,15,4,3])
print(j)
j = np.unique(j)
print(j)
```

```python

print(mt_bidimensional)
print(mt_bidimensional[0][1])
print(mt_bidimensional.shape)
```

```python
#Funções matemáticas
#retorna maior valor da matriz
print(mt_bidimensional.max())
#retorna o menor valor da matriz
print(mt_bidimensional.min())
#retorna soma dos valores da matriz
print(mt_bidimensional.sum())
#retorna a média dos valores da matriz
print(mt_bidimensional.mean())
#desvio padrão (standard deviation) dos valores da matriz
print(mt_bidimensional.std())
```

```python
#aplicando funções a todos itens da matriz
k1 = np.array([4,16,36,9,81])
print(np.sqrt(k1)) #raiz quadrada
print(np.exp(k1)) #valor exponencial
```

```python
#extracao de elementos
#elemento da posicao 2
print(k1[1])
#Mostra a partir da posição n, 2 elementos
n=1
print(k1[n:n+2])
#mostra o array da terceira posição em diante
print(k1[2:])
#mostra o array da penúltima posição até o final
print(k1[-2:])
#mostra o array da penúltima posição até a primeira
print(k1[-2::-1])
```

```python
#extração de linhas e colunas
l = np.array([[2,3],[4,5],[6,1],[7,8]])
print(l)
#primeira linha, todas as colunas
l_linha_1 = l[0,:]
print(f"Primeira linha: {l_linha_1}")
#segunda linha
l_linha_2 = l[1,:]
print(f"Segunda linha: {l_linha_2}")
#terceira linha
l_linha_3 = l[2,:]
print(f"Terceira linha {l_linha_3}")
l_linha_4 = l[3,:]
print(f"Quarta linha {l_linha_4}")

```

```python
#todas as linhas, primeira coluna
l_coluna_1 = l[:,0]
print(f"Primeira coluna {l_coluna_1}")
l_coluna_2 = l[:,1]
print(f"Segunda coluna {l_coluna_2}")
```

```python
#adição e multiplicação de matrizes
n = np.array([[2,4],[6,8]])
o = np.array([[1,3],[5,7]])
res1 = n + o
print(n,'\n',o,'\nSoma de matrizes:')
print(res1)
print('Multiplicação de matrizes')
res2 = n * o
print(res2,'\n')

p = np.array([[2,4],[8,9],[10,11],[16,23]])
q = np.array([2,1])
print(p+q)

```

```python
#transpoisção, rearranja um conjunto de 15 elementos de 0 a 14
f = np.arange(15).reshape((3,5))
print(f)
print('matriz transposta')
s = f.T
print(s)
```

```python
#expressoes logicas usando where
#criando matriz com valores aleatorios positivos e negativos
mt_ale = np.random.randn(4,4)
print(mt_ale)
#criando matriz com valores booleanos
validador = (mt_ale > 0)
print(validador)
#criando matriz com valores 'positivo' ou 'negativo' baseado nas condições da 
#array validador
p_n = np.where(validador>0,'Positivo','Negativo')
print(p_n)
```
