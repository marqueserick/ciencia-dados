```python
#faça um programa que tenha uma funcao chamada amplitude
#a funcao deve receber uma lista e imprimir a amplitude
#amplitude é a diferença do maior e menor valor
def amplitude(lista):
  amplitude = max(lista) - min(lista)
  print(f"Amplitude da lista: {amplitude}")

lst = [21,22,23,24,150]
amplitude(lst)
```

```python
#faça um funcao que receba uma string e imprime essa string em forma vertical
def imprimir_vertical(texto):
  for i in range(0,len(texto)):
    print(texto[i])

nome = 'Erick Marques'
imprimir_vertical(nome)
```

```python
#crie um programa que leia o peso de uma carga em numeros inteiros. Se o peso for
#até 10kg, informe que o valor será de R$ 50,00. Entre 11 e 20 kg, o valor será 
#de 80,00. Se for maior que 20, informe que o transporte não é aceito. Teste vários valores.
def verifica_peso(peso):
  if peso <= 10:
    print("Valor do transporte: R$50,00")
  elif peso>10 and peso<=20:
    print("Valor do transporte: R$80,00")
  else:
    print("Transporte não é aceito para esse peso")

pesos = [10,11,23,5,8,19,1]
for peso in pesos:
  print(f"Peso: {peso:.2f}kg")
  verifica_peso(peso)
```
