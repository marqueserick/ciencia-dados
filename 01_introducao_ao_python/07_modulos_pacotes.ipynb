```python
#conjunto de funcionalidades organizadas em arquivos
#usamos import para acessar os arquivos
# palavra reservada 'as' dá um apelido para o import
#apos importar podemos usar as funcoes do modulo

import statistics as est
z = [6,5,9,8,3]
x = est.median(z)
y = est.mean(z)
print(f"Mediana {x}\nMédia {y}")
```

```python
#importar funcoes especificas
from statistics import median,mean
z = [10,20,30,40]
x = mean(z)
print(f"média: {x}")
y = median(z)
print(f"mediana: {y}")
```
