```python
#Repetir um bloco de código numa quantidade de vezes,geralmente, determinada de forma dinamica
#while executa enquanto uma condição for verdadeira
#for executa um n numero de vezes
a = 0
print("Estrutura de repetição while:")
while a<10:
  print(a)
  a+=1

print("Estrutura de repetição for:")
for a in range(0,10):
  print(a)
```

```python
#interrupções: break e continue
#break cancela totalmente o laço
#continue pula a interação atual do laço

print("Números menor que 6")
for i in range (1,10):
  if  i>5:
    break
  print(i)

print("Números pares")
for i in range(1,10):
  if i%2 == 1:
    continue
  print(i)
```
